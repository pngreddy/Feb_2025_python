**Step-11: Drop duplicates**

```python
dict1={'A':[1,2,3,4,4,4],
       'B':['Apple','Banana','Cherry','Banana','Bread','Bread']}
df2=pd.DataFrame(dict1)
df2
```

```python
df2.drop_duplicates(subset=['A'])
# drop the repeated rows and keep at lease one
```

```python
df2
```

```python
df2.drop_duplicates(subset=['A'],
                   keep='last')
```
