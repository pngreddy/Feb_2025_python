**Outliers**

- Outliers are the observation which skewed the data

- Outliers either very very huge values or very very less values

- Outliers affect the mean

- If outliers are presesnt the mean will not equal to median

```python
wage_data=visa_df['prevailing_wage']
wage_data.mean()==wage_data.median()
```
