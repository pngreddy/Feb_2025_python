**Bar chart**

- in matplotlib we have a method plt.bar

- we know that in order to get bar plot need two columns

  - categorical column which indicates lables

  - numerical column which indicates corresponding values

- bar chart we will get from frquency table

- from there we extract labels and values

- value_counts will give lables and values

```python
labels=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
plt.figure(figsize=(10,6)) # 10 untis x-axis, 6 units y-axis
plt.bar(labels,values)
plt.xlabel('Labels') # X-axis name
plt.ylabel('Count') # Y-axis name
plt.title('Continent_Bar_chart') # Title of the plot
plt.savefig('Continent_Bar_chart.jpg')
plt.show()  # write up will not come, show always at last
```
