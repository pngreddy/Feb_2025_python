## Encoding

- Encoding is a method to convert Categorical column data to Numerical

- it is very important step, becuase ML algorithms developed based on Maths

- Maths need numerical values

    - Map

    - LabelEncoder

    - OneHotEncoder

    - np.where 

**Map**

- Map method means mapping the categorical labels to numerical

- For example take case status column

- It has two unique lables

    - Certifed

    - Denied

- Now I want to map Certified with 0 and Denied with 1

- We are following alphabetical order and python index start with zero

- we need to create a dictionary lables as keys , the numbers as values

```python
visa_df['case_status'].unique()
```

```python
dict1={'Certified':0,'Denied':1}
visa_df['case_status']=visa_df['case_status'].map(dict1)
```
